**3. Veri Setinde Eksik Değer Kontrolü**

```python
print("=== Eksik Değer Kontrolü ===")

print("\nORDERS Missing Values:")
print(orders.isnull().sum())

print("\nCUSTOMERS Missing Values:")
print(customers.isnull().sum())

print("\nPRODUCTS Missing Values:")
print(products.isnull().sum())
```

**Eksik Müşteri Bilgilerini Düzeltme**

```python
# Eksik değer içeren müşteri satırlarını göster
missing_customers = customers[customers.isnull().any(axis=1)]
print(missing_customers)
#A) Eksik müşteri bilgisi olan satırları tespit etme
```

```python
#Eksik müşteri bilgilerini doldurma
#Genelde müşteri bilgisi eksikse:
#email → boş ise: first_name.last_name@example.com
#şeklinde doldur
#first_name / last_name → eksikse ‘Unknown’
#signup_date → ortalama tarih ile doldurulabilir
# first_name veya last_name boşsa
customers['first_name'] = customers['first_name'].fillna("Unknown")
customers['last_name'] = customers['last_name'].fillna("Unknown")

# email boşsa otomatik email oluşturma
customers['email'] = customers['email'].fillna(
    customers['first_name'].str.lower() + "." +
    customers['last_name'].str.lower() + "@example.com"
)

# signup_date eksikse ortalama tarih ile doldurma
customers['signup_date'] = customers['signup_date'].fillna(
    customers['signup_date'].mode()[0]
)
```

```python
#Alternatif: Eksik müşteri bilgileri olan satırları tamamen silme
customers_cleaned = customers.dropna()
```

```python
#2. Ürün Fiyatı Eksik Olan Ürünleri Belirleme
missing_price_products = products[products['price'].isnull()]
print("Fiyatı eksik ürünler:")
print(missing_price_products)
```

```python
#A) Şirket politikası 1: Eksik fiyatları ortalama fiyat ile doldur
ortalamaFiyat = products['price'].mean()
products['price'] = products['price'].fillna(ortalamaFiyat)
print("Fiyatlar dolduruldu.")
print(products.head())
```

```python
#B) Şirket politikası 2: Fiyatı eksik ürünleri tamamen sil
products_cleaned = products.dropna(subset=['price'])
```

```python
#BONUS → Ürün kategorisine göre ortalama fiyat ile doldurma
products['price'] = products.groupby('category')['price'].transform(
    lambda x: x.fillna(x.mean())
)
```

**Tarih Sütunlarını datetime Tipine Çevirme**                      
Genelde aşağıdaki sütunlar tarih içerir:               
orders.order_date              
customers.signup_date                       

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
customers['signup_date'] = pd.to_datetime(customers['signup_date'])
```

**Birden fazla tarih sütununu otomatik tespit edip dönüştürme:**

```python
# Tarih içeren kolon adlarında "date" geçenleri otomatik datetime yapma
for col in orders.columns:
    if "date" in col:
        orders[col] = pd.to_datetime(orders[col])

for col in customers.columns:
    if "date" in col:
        customers[col] = pd.to_datetime(customers[col])
```

**Sayısal Olması Gereken Sütunları Kontrol Etme**                      
Sayısal kategoride olması gereken sütunlar:                      
orders.quantity                    
orders.price                   
roducts.price                       
products.stock                    

```python
#Bu sütunların veri tipini kontrol etme:
print("Orders veri tipleri:")
print(orders.dtypes)

print("\nProducts veri tipleri:")
print(products.dtypes)

print("\nCustomers veri tipleri:")
print(customers.dtypes)
```

**Sayısal Olmayan Değerleri Sayısal Tipe Zorlama (gerekirse)**              
Eğer veride yanlışlıkla string bir değer varsa (ör. "1", "5 adet", "NaN") bunları sayıya dönüştürme:

```python
# Orders tablosu
orders['quantity'] = pd.to_numeric(orders['quantity'], errors='coerce')
orders['price'] = pd.to_numeric(orders['price'], errors='coerce')

# Products tablosu
products['price'] = pd.to_numeric(products['price'], errors='coerce')
products['stock'] = pd.to_numeric(products['stock'], errors='coerce')
```

**Sayısal olmayan değerleri listeleme (temizlik için çok faydalı)**   
Hangi satırda problem olduğunu görmek için:

```python
invalid_price = products[pd.to_numeric(products['price'], errors='coerce').isnull()]
print("Sayısal olmayan fiyatlar:", invalid_price)

invalid_quantity = orders[pd.to_numeric(orders['quantity'], errors='coerce').isnull()]
print("Sayısal olmayan adetler:", invalid_quantity)
```

Siparişleri Müşteri Bilgileri ile Birleştirme (merge)             
Amaç:             
orders + customers → order_customer                

```python
#Birleştirme anahtarı:
order_customer = orders.merge(customers, on="customer_id", how="left")
#how="left": Tüm siparişler korunur, müşteri bilgisi yoksa NaN kalır (genelde e-ticarette doğru yöntem).
```

 **Müşteri + Sipariş Tablosunu Ürün Bilgileri ile Birleştirme**             
Amaç: order_customer + products → full_data

```python
#Birleştirme anahtarı: product_id
full_data = order_customer.merge(products, on="product_id", how="left")
```

**Sonuç: Tek bir analiz tablon var → full_data**
| Tür                      | Kaynağı   |
| ------------------------ | --------- |
| Sipariş bilgileri        | orders    |
| Müşteri bilgileri        | customers |
| Ürün bilgileri           | products  |
| Tümleşik analiz alanları | full_data |


**Tüm işlemleri tek blok kod olarak görmek istersek:**

```python
# 1. Orders + Customers
order_customer = orders.merge(customers, on="customer_id", how="left")

# 2. (Orders + Customers) + Products
full_data = order_customer.merge(products, on="product_id", how="left")

# Sonuç
print(full_data.head())
```

**En Çok Satılan İlk 10 Ürün**      
Satış adedi = toplam sipariş miktarı (quantity)           

```python
top_10_products = (
    full_data.groupby("product_name")["quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_10_products)
```

**En Yüksek Gelir Getiren Kategori**              
Gelir = price × quantity                
Önce yeni bir gelir sütunu oluşturalım:

```python
full_data["revenue"] = full_data["price"] * full_data["quantity"]
```

```python
#Sonra kategori bazında gelir hesaplayalım:
category_revenue = (
    full_data.groupby("category")["revenue"]
    .sum()
    .sort_values(ascending=False)
)

print(category_revenue)
```

```python
top_category = category_revenue.idxmax()
print("En yüksek gelir getiren kategori:", top_category)
```

**Aylara Göre Toplam Satış Geliri (Line Chart İçin Hazırlık)**    
Ay bilgisini çıkart:

```python
full_data["order_month"] = full_data["order_date"].dt.to_period("M")
```

```python
#Aylık gelir hesaplama:
monthly_revenue = (
    full_data.groupby("order_month")["revenue"]
    .sum()
    .sort_values()
)

print(monthly_revenue)
```

```python
#Matplotlib ile line chart çizmek istersen:
import matplotlib.pyplot as plt

monthly_revenue.plot(kind="line", marker="o")
plt.title("Aylara Göre Toplam Satış Geliri")
plt.xlabel("Ay")
plt.ylabel("Gelir")
plt.grid(True)
plt.show()
```

**Sipariş Başına Ortalama Gelir**              
Her siparişin toplam gelirini hesaplayalım:

```python
order_revenue = full_data.groupby("order_id")["revenue"].sum()
```

```python
#Ortalama sipariş geliri:
avg_order_revenue = order_revenue.mean()
print("Sipariş başına ortalama gelir:", avg_order_revenue)
```

**En Fazla Alışveriş Yapan Müşteri**     
"En fazla alışveriş" → en çok sipariş adedi veren müşteri

```python
top_customer = (
    full_data.groupby(["customer_id", "first_name", "last_name"])["order_id"]
    .nunique()
    .sort_values(ascending=False)
    .head(1)
)

print(top_customer)
```

```python
#Eğer adet (quantity) bazlı en çok ürün satın alan müşteri hangisi:
top_customer_by_quantity = (
    full_data.groupby(["customer_id", "first_name", "last_name"])["quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(1)
)

print(top_customer_by_quantity)
```

**Müşterileri Yıllık Harcama Tutarına Göre Sınıflandırma**    
Önce yıllık harcama tutarını hesaplayalım:

```python
full_data["revenue"] = full_data["price"] * full_data["quantity"]

customer_spend = full_data.groupby("customer_id")["revenue"].sum()
```

**Harcama kategorileri oluşturma (örnek politika)**               
Bronze: 0 – 2,000 TL          
Silver: 2,000 – 5,000 TL                  
Gold: 5,000+ TL

```python
customer_segments = pd.cut(
    customer_spend,
    bins=[0, 2000, 5000, float("inf")],
    labels=["Bronze", "Silver", "Gold"]
)

customer_segments = customer_segments.reset_index()
customer_segments.columns = ["customer_id", "segment"]

print(customer_segments)
```

**En Çok Gelir Getiren Şehir (city Sütunu Üzerinden)**

```python
# Örnek: customers['city'] sütunu mevcut olsun
#Önce sipariş + müşteri + ürün birleşik tabloyu şehir bilgisiyle genişletiyoruz:
full_with_city = full_data.merge(customers[['customer_id', 'city']], on='customer_id', how='left')
```

```python
#Sonra şehir bazında gelir hesaplıyoruz:
city_revenue = (
    full_with_city.groupby("city")["revenue"]
    .sum()
    .sort_values(ascending=False)
)

print(city_revenue)
```

```python
#En çok gelir getiren şehir:
top_city = city_revenue.idxmax()
print("En çok gelir getiren şehir:", top_city)
```

**Önce tarih formatı uygunsa direkt gün bilgisi çıkarılabilir:**

```python
full_data["order_date"] = pd.to_datetime(full_data["order_date"])
```

**Gün bazında sipariş adedi:**

```python
orders_by_day = (
    full_data.groupby(full_data["order_date"].dt.date)["order_id"]
    .nunique()
)

print(orders_by_day)
```

```python
#Eğer aynı gün verilen birleşik siparişler tekrar ediyorsa, quantity toplamı isteniyorsa:
orders_by_day_quantity = (
    full_data.groupby(full_data["order_date"].dt.date)["quantity"]
    .sum()
)

print(orders_by_day_quantity)
```

**Haftanın Hangi Günü Daha Çok Satış Olmuş?**

```python
full_data["day_of_week"] = full_data["order_date"].dt.day_name()
```

```python
#Haftanın gününe göre sipariş adedi:
orders_by_weekday = (
    full_data.groupby("day_of_week")["order_id"]
    .nunique()
    .sort_values(ascending=False)
)

print(orders_by_weekday)
```

```python
#Haftanın gününe göre toplam satış miktarı (adet):
quantity_by_weekday = (
    full_data.groupby("day_of_week")["quantity"]
    .sum()
    .sort_values(ascending=False)
)

print(quantity_by_weekday)
```

```python
#En çok sipariş gelen günü bulma
top_day = orders_by_weekday.idxmax()
print("En çok sipariş gelen gün:", top_day)
```

```python
#Eğer adet bazlı istersen:
top_day_quantity = quantity_by_weekday.idxmax()
print("En çok ürün satılan gün:", top_day_quantity)
```
